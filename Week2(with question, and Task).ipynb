5. Create 5 boxes of different sizes and colors in one image. it is recommended to use 
random numbers!

```python
from random import randint

# make 2d geometry starting with black image
black_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)

# loop 5 iterations to draw rectangle with random width & height points, random colors and thickness
for _ in range(0, 5):
  cv.rectangle(black_img, pt1=(randint(0, 512),randint(0, 512)), pt2=(randint(0, 512),randint(0, 512)), color=(randint(0, 255),randint(0, 255),randint(0, 255)), thickness=randint(5,15))

plt.imshow(black_img)
```
